```python
# import library
import xarray as xr
import hvplot.xarray
import hvplot.pandas
import rasterio as rio
import rioxarray
from matplotlib import pyplot as plt
```

```python
# Get image and plot
topo_url = (
    "https://www.landsd.gov.hk/landsd_psi_data/SMO/image/B200K_R500index-geo.tif")
topo_data = xr.open_dataset(topo_url, engine="rasterio")

topo_map = topo_data.hvplot.image('x', 'y', width=700, height=500, cmap="copper", geo=True)
```

```python
with rio.open(topo_url) as src:
    # Read the data from the first band
    data = src.read(1)

    # Create a figure and axis for the plot
    fig, ax = plt.subplots(figsize=(8, 6))

    # Plot the data
    # The 'cmap' can be adjusted for different color maps
    cax = ax.imshow(data, cmap='copper')

    # Add a colorbar
    fig.colorbar(cax, ax=ax)

    # Set title and labels if desired
    ax.set_title('Topography Map from TIFF Data')
    ax.set_xlabel('X coordinate')
    ax.set_ylabel('Y coordinate')

    # Show the plot
    plt.show()
```
